### --- Task 1: Load and Explore the Dataset ---

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = r'C:\Users\User\Documents\plp\Projects\sales_data.csv'
df = pd.read_csv(file_path)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Display first few rows
display(df.head())
```

### EDA

```python
# Dataset info
df.info()
```

```python
# Missing values check
df.isnull().sum()
```

### --- Task 2: Basic Data Analysis ---

```python
# Descriptive statistics
desc_stats = df.describe()
display(desc_stats)
```

```python
# Average revenue per product
avg_revenue_per_product = df.groupby("Product")["Revenue ($)"].mean()
display(avg_revenue_per_product)
```

### --- Task 3: Data Visualization ---

```python
# 1. Bar chart: Average revenue per product
plt.figure(figsize=(8,4))
avg_revenue_per_product.plot(kind='bar', color='skyblue')
plt.title("Average Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Average Revenue ($)")
plt.show()
```

```python
# 2. Histogram: Distribution of Quantity Sold
plt.figure(figsize=(8,4))
df["Quantity Sold"].hist(bins=10, color='orange', edgecolor='black')
plt.title("Distribution of Quantity Sold")
plt.xlabel("Quantity Sold")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Scatter plot: Quantity Sold vs Revenue
plt.figure(figsize=(8,4))
plt.scatter(df["Quantity Sold"], df["Revenue ($)"] , alpha=0.6)
plt.title("Quantity Sold vs Revenue")
plt.xlabel("Quantity Sold")
plt.ylabel("Revenue ($)")
plt.show()
```

### --- Findings and Observations ---

##### 1. The dataset covers multiple products (Laptop, Mouse, Keyboard).
##### 2. Revenue trends vary across dates, with some spikes in sales.
##### 3. Laptops generate high revenue even when quantities are low (suggesting high unit price).
##### 4. Distribution of quantities shows that most sales involve smaller quantities.
##### 5. Scatter plot shows a weak relationship between quantity sold and revenue, influenced by product type.
